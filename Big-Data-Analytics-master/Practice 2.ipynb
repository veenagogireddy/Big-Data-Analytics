## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
# File location and type
file_location = "/FileStore/tables/gun_violence_data_01_2013_03_2018-edcc2.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
# Create a view or table

temp_table_name = "gunviolencecsv"

df.createOrReplaceTempView(temp_table_name)
```

```
df = sqlContext.sql("select * from gunviolencecsv")
display(df)
df.count()
df.show()
```

```
df=df.select('*', (df.n_killed + df.n_injured).alias('n_total'))
```

```
df.show()
```

```
df=df.na.drop()
```

```
df3=df.select('latitude','longitude')
```

```
from pyspark.sql.types import *
df3 = df3.withColumn("latitude", df["latitude"].cast(FloatType())) \
.withColumn("longitude", df["longitude"].cast(FloatType()))
```

```
from pyspark.ml.feature import Imputer
imputer = Imputer(inputCols=["latitude", "longitude"], outputCols=["out_a", "out_b"])
model = imputer.fit(df3)
df4=model.transform(df3)
```

```
df4.printSchema()
```

```
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler

assembler = VectorAssembler(
    inputCols=["out_a","out_b"],
    outputCol="features")
output = assembler.transform(df4)
```

```
output.printSchema()
```

```
df4 = output.selectExpr("features as features")
```

```
df4.printSchema()
```

```
from pyspark.ml.clustering import KMeans

# Loads data.
#dataset = spark.read.format("libsvm").load("data/mllib/sample_kmeans_data.txt")

# Trains a k-means model.
kmeans = KMeans().setK(2).setSeed(1)
model = kmeans.fit(df4)

# Evaluate clustering by computing Within Set Sum of Squared Errors.
wssse = model.computeCost(df4)
print("Within Set Sum of Squared Errors = " + str(wssse))

# Shows the result.
centers = model.clusterCenters()
print("Cluster Centers: ")
for center in centers:
    print(center)
```
